```python


import os
from mistralai.client import Mistral
from google.colab import userdata

# SECURELY fetch the API key from Colab's Secrets manager
try:
    MISTRAL_API_KEY = userdata.get('MISTRAL_API_KEY')
except userdata.SecretNotFoundError:
    raise ValueError("API Key not found! Please add MISTRAL_API_KEY to the Secrets tab (key icon on the left).")

# Initialize the Mistral client
client = Mistral(api_key=MISTRAL_API_KEY)

def ask_health_bot_mistral(user_query):
    system_prompt = """
    You are a friendly, empathetic, and highly knowledgeable medical assistant.
    Your objective is to answer general health queries clearly and simply.

    CRITICAL SAFETY RULES:
    1. You are an AI, not a licensed doctor. You must NEVER diagnose conditions, prescribe medications, or offer definitive medical treatment plans.
    2. Always include a brief, gentle disclaimer reminding the user to consult a healthcare professional for personalized advice.
    3. If a prompt mentions severe pain, difficulty breathing, or a potential emergency, instruct the user to seek immediate emergency medical care.
    """

    try:
        # Calling the Mistral API
        response = client.chat.complete(
            model="open-mistral-7b",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_query}
            ],
            temperature=0.2,
            max_tokens=300
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"Connection Error: {e}"

# ==========================================
# Testing the health_Bot
# ==========================================
print("Initializing Mistral Health Bot...\n" + "="*50)

queries = [
    "What causes a sore throat?",
    "Is paracetamol safe for children?"
]

for q in queries:
    print(f"\n🗣️ USER: {q}")
    print(f"🤖 HEALTHBOT:\n{ask_health_bot_mistral(q)}")
    print("-" * 50)
```

```python
# Install the mistral library first if you haven't already
!pip install mistralai -q

import os
from mistralai.client import Mistral
from google.colab import userdata

# SECURELY fetch the API key from Colab's Secrets manager
try:
    MISTRAL_API_KEY = userdata.get('MISTRAL_API_KEY')
except userdata.SecretNotFoundError:
    raise ValueError("API Key not found! Please add MISTRAL_API_KEY to the Secrets tab (key icon on the left).")

# Initialize the Mistral client
client = Mistral(api_key=MISTRAL_API_KEY)

def ask_health_bot_mistral(user_query):
    system_prompt = """
    You are a friendly, empathetic, and highly knowledgeable medical assistant.
    Your objective is to answer general health queries clearly and simply.

    CRITICAL SAFETY RULES:
    1. You are an AI, not a licensed doctor. You must NEVER diagnose conditions, prescribe medications, or offer definitive medical treatment plans.
    2. Always include a brief, gentle disclaimer reminding the user to consult a healthcare professional for personalized advice.
    3. If a prompt mentions severe pain, difficulty breathing, or a potential emergency, instruct the user to seek immediate emergency medical care.
    """

    try:
        # Calling the Mistral API
        response = client.chat.complete(
            model="open-mistral-7b",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_query}
            ],
            temperature=0.2,
            max_tokens=300
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"Connection Error: {e}"

# ==========================================
# Interactive Chat Loop
# ==========================================
print("==================================================")
print("🤖 Mistral Health Bot Initialized!")
print("Type your health questions below.")
print("Type 'quit', 'exit', or 'q' to stop the chat.")
print("==================================================")

while True:
    # Prompt the user for input
    user_input = input("\n🗣️ YOU: ")

    # Check if the user wants to exit
    if user_input.strip().lower() in ['quit', 'exit', 'q']:
        print("\n🤖 HEALTHBOT: Take care and stay healthy! Goodbye!")
        break

    # Skip empty questions
    if not user_input.strip():
        continue

    # Get and print the bot's response
    bot_response = ask_health_bot_mistral(user_input)
    print(f"🤖 HEALTHBOT:\n{bot_response}")
    print("-" * 50)
```
